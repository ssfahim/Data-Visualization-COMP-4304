```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import ticker
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv('population.csv')
```

## Question 1


```python
df1 = df[(df.Country == 'Canada')]
x = df1.Year
y = df1.Population
plt.plot(x,y)

def axisLabel(x, pos):
    if x>= 1e6:
        return'%3d M'%(x*1e-6)
    else:
        return x
plt.gca().yaxis.set_major_formatter(axisLabel)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title("Canada's Population Growth")
```

## Question 2

```python
df_population = df[['Population', 'UrbanPopulation','Year']].groupby('Year').sum().reset_index().rename(columns={'Population': 'total_Population', 'UrbanPopulation':'total_UrbanPopulation'})
df_population['Urban_percentage'] = (df_population.total_UrbanPopulation/df_population.total_Population)*100

x = df_population.Year
y = df_population.Urban_percentage
plt.plot(x,y)
# I got this piece of code from the below mentioned site
# https://stackoverflow.com/questions/31357611/format-y-axis-as-percent
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=100.0, decimals=0))

plt.ylim(bottom = 30)
plt.xlabel('Year')
plt.ylabel('Urban Percentage')
plt.title('Percentage of Global Population living in Urban Areas')
```
